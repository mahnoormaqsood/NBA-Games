```python
##programme that specifically catches the extra box error for file "nbagame0021400022" 
##the script can be edited by saving the starting value for each quarter in a variable rather that using 720 for 
##other games that may not have the quarter starting value at 720
##Some code reproduced from 'http://savvastjortjoglou.com/nba-play-by-play-movements.html'
##August 2016

import json
import turtle as t
import pandas as pd
import numpy as np
import os
import re
from tkinter import * 



filename = "nbagame0021400041.json"
length = 50
track1 = 0
track2 = 0
track3 = 0
track4 = 0

events = []
with open(filename, 'r') as f:
    for line in f:
        events.append(json.loads(line))
        for i in events:
            pass 

# Column labels|
headers = ["team_id", "player_id", "x_loc", "y_loc", 

           "radius", "moment", "game_clock", "shot_clock"]


# Write text at location (x, y)
def writeText(s, x, y): 
    ttl.penup() # Pull the pen up
    ttl.goto(x, y)
    ttl.pendown() # Pull the pen down
    ttl.write(s, font=("Times",20,"bold")) # Write a string


# setup screen size
t.setup (1000, 800, 0, 0)
# create a turtle object
ttl = t.Turtle()
ttl.penup()
ttl.goto(-400, -400)
ttl.setheading(90)
ttl.pendown()
ttl.forward(800)
ttl.penup()
ttl.goto(-450, -350)
ttl.setheading(0)
ttl.pendown()
ttl.forward(1000)
p = 400

writeText('0', 500, -370)
writeText('Game Clock Countdown', 0, -390)
writeText('720', -220, -370)
writeText('360', 140, -370)
writeText('Quarter 1', -520, 300)
writeText('Quarter 2', -520, 100)
writeText('Quarter 3', -520, -150)
writeText('Quarter 4', -520, -300)
writeText(filename, -600, -50)
flag = True


def drawBox(ttl, x1, x2):
    ttl.penup()
    ttl.goto(500 - x2, p)
    #print('VALUES:')
    #print('x2 is: %f', x2)
    #print('x1 is: %f', x1)
    ttl.setheading(0)	# set the pen in the +ve x direction
    ttl.pendown()
    ttl.forward(x2-x1)
    ttl.right(90)
    ttl.forward(length)
    ttl.right(90)
    ttl.forward(x2-x1)
    ttl.right(90)
    ttl.forward(length)
    ttl.penup()


c = 0
for event in events:
    moments = event['moments']
    gameid = event['gameid']
    home = event['home']
    visitor = event['visitor']
    gamedate = event['gamedate']

    # Initialize our new list
    player_moments = []
    quarter = []

    for moment in moments:  
        #if moment[0] == 1:
         #   track = track + 1
        # For each player/ball in the list found within each moment
        for player in moment[5]:
            # Add additional information to each player/ball
            # This info includes the index of each moment, the game clock
            # and shot clock values for each moment
            player.extend((moments.index(moment), moment[2], moment[3]))
            player_moments.append(player)
        for q in str(moment[0]):
            quarter.append(q)

    
    # inspect our list
    #print('printing:')        
    #player_moments[0:11]
    df = pd.DataFrame(player_moments, columns=headers)
    #df.head(11)
    # creates the players list with the home players
    players = home["players"]
    # Then add on the visiting players
    players.extend(visitor["players"])
    # initialize new dictionary
    id_dict = {}

    # Add the values we want
    for player in players:
        id_dict[player['playerid']] = [player["firstname"]+" "+player["lastname"],player["jersey"]]
    id_dict.update({-1: ['ball', np.nan]})
    df["player_name"] = df.player_id.map(lambda x: id_dict[x][0])
    df["player_jersey"] = df.player_id.map(lambda x: id_dict[x][1])
    ttl.color('blue')
    

    if quarter[0] == '1':
      
        if player_moments[len(player_moments) - 1][6] == 720:
           
            if track1 == 0:
               
                drawBox(ttl,player_moments[len(player_moments) - 1][6], player_moments[0][6])
            else:
               
                pass

        else:
           
            drawBox(ttl,player_moments[len(player_moments) - 1][6], player_moments[0][6])
        track1 = track1 + 1
    
    elif quarter[0] == '2':
       
        if player_moments[len(player_moments) - 1][6] == 720:
            
            if track2 == 0:
               
                drawBox(ttl,player_moments[len(player_moments) - 1][6], player_moments[0][6])
            else:
               
                pass
        else:
       
            drawBox(ttl,player_moments[len(player_moments) - 1][6], player_moments[0][6])
        track2 = track2 + 1
        
    elif quarter[0] == '3':
       
        if player_moments[0][6] == 720.0:
            print('CP14')
            print('HERE')
            print(player_moments[len(player_moments) - 1][6])
            print(player_moments[0][6])
            if track3 == 0:
              
                drawBox(ttl,player_moments[len(player_moments) - 1][6], player_moments[0][6])
            else:
            
                pass
        else:
            
            drawBox(ttl,player_moments[len(player_moments) - 1][6], player_moments[0][6])
        track3 = track3 + 1
    else:
        if player_moments[len(player_moments) - 1][6] == 720:
            if track4 == 0:
                drawBox(ttl,player_moments[len(player_moments) - 1][6], player_moments[0][6])
            else:
                pass
        else:
            drawBox(ttl,player_moments[len(player_moments) - 1][6], player_moments[0][6])
        track4 = track4 + 1
    
          
            
    p = p - 1.5
    c = c + 1



t.done()


print(c)









```
